```python
path='/home/octo/Jts/C:\Jts\TradeReport'+ '/trades.20180423.csv'
```

```python
trade_report=pd.read_csv(path,
            parse_dates={'time': ['Date','Time']},
            index_col='time',
            usecols=['Action', 'Quantity', 'Symbol', 'Price','Time','Date','Realized P&L'])
```

```python
trades=trade_report
```

```python
trade_report.Quantity=np.where(trade_report.Action=='SLD',trade_report.Quantity*-1,trade_report.Quantity)
```

```python
trade_report=trade_report.pivot_table(index='time',columns='Action',aggfunc=sum).drop('Price',axis=1)
```

```python
trade_report.fillna(0,inplace=True)
```

```python
trade_report['holding']=(trade_report.Quantity.BOT+trade_report.Quantity.SLD)
```

```python
trade_report['realizedPL']=(trade_report['Realized P&L'].BOT+trade_report['Realized P&L'].SLD)
```

```python
trade=pd.merge(pd.concat([trade_report.holding,trade_report.realizedPL], axis=1),trades[['Price']],right_index=True, left_index=True)
```
