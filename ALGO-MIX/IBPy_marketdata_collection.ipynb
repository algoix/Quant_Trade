Market Data from IB by R in .dat format.Need to import for analysis

```python
import pandas as pd   
import numpy as np

with open('/home/octo/Desktop/PROJECT_1/data_R/SPY.dat','r') as f:
    #next(f) # skip first row
    df = pd.DataFrame(l.rstrip().split() for l in f)
```

```python
df['time'] = df[0].str.cat(df[1].values.astype(str), sep=' ')
```

```python
df.index=df['time']
df=df[[4,5,6,7]]
```

#option call/put volume and interest is avialable in the dataset

```python
dfBID=df[df[4]=='bidPrice:']
dfASK=df[df[4]=='askPrice:']
dfV=df[df[4]=='Volume:']
dfAV=df[df[4]=='averageVolume:']
dfCOI=df[df[4]=='optionCallOpenInterest:']
dfPOI=df[df[4]=='optionPutOpenInterest:']
dfCV=df[df[4]=='optionCallVolume:']
dfCV=df[df[4]=='optionPutVolume:']
dfIV=df[df[4]=='optionImpliedVol:']
```
